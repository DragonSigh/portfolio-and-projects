```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
netflix_df = pd.read_csv("netflix_data.csv")
netflix_df.head()
```

```python
# Subset the DataFrame for "Movies"
netflix_subset = netflix_df[netflix_df["type"] == "Movie"]

# Subset the columns of the new DataFrame
netflix_movies = netflix_subset[["title", "country", "genre", "release_year", "duration"]]

# Filter the DataFrame by Movie duration
short_movies = netflix_movies[netflix_movies["duration"] < 60]
```

```python
# Get all unique genres for movies
netflix_movies["genre"].unique()
```

```python
# Assign colors to movie genres
colors = []
for lab, row in netflix_movies.iterrows():
    if row["genre"] == "Dramas":
        colors.append("blue")
    elif row["genre"] == "Horror Movies":
        colors.append("black")
    elif row["genre"] == "Action":
        colors.append("red")
    elif row["genre"] == "Comedies":
        colors.append("yellow")
    elif row["genre"] == "Children":
        colors.append("green")
    elif row["genre"] == "Romantic Movies":
        colors.append("pink")
    else:
        colors.append("gray")
```

```python
# Plot the movie duration by year using the genre colors created
fig = plt.figure(figsize=(12, 8))
plt.scatter(netflix_movies["release_year"], netflix_movies["duration"], c=colors)
plt.xlabel("Release year")
plt.ylabel("Duration (min)")
plt.title("Movie Duration by Year of Release")
```
